# This code generates a file structure

```python
from pathlib import Path

# =========================================
# CONFIG
# =========================================
ROOT_DIR = Path(".")

EXCLUDE_DIRS = {
    "__pycache__",
    ".git",
    "node_modules",
    "image"          # hides dataset image contents
}

EXCLUDE_EXTENSIONS = {
    ".jpg",
    ".jpeg",
    ".png",
    ".webp",
    ".gif",
    ".pyc",
    ".txt"
}

# =========================================
# TREE GENERATOR
# =========================================
def generate_tree(path: Path, prefix=""):
    entries = sorted(
        [p for p in path.iterdir()
         if p.name not in EXCLUDE_DIRS
         and p.suffix.lower() not in EXCLUDE_EXTENSIONS],
        key=lambda x: (x.is_file(), x.name.lower())
    )

    for i, entry in enumerate(entries):
        connector = "в””в”Ђв”Ђ " if i == len(entries) - 1 else "в”њв”Ђв”Ђ "

        print(prefix + connector + entry.name)

        if entry.is_dir():
            extension = "    " if i == len(entries) - 1 else "в”‚   "
            generate_tree(entry, prefix + extension)

# =========================================
# RUN
# =========================================
print(ROOT_DIR.resolve().name)
generate_tree(ROOT_DIR)
```
